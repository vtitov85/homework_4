```python
import pandas as pd
import numpy as np
```

```python
school_data_to_load = "Desktop/schools_complete.csv"
student_data_to_load = "Desktop/students_complete.csv"
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
total_schools = school_data_complete['school_name'].nunique()
```

```python
total_schools
```

```python
student_count = school_data_complete['student_name'].count()
```

```python
student_count
```

```python
total_budget = school_data['budget'].sum()
```

```python
total_budget
```

```python
avg_math_score = school_data_complete['math_score'].mean()
```

```python
avg_math_score
```

```python
avg_reading_score = school_data_complete['reading_score'].mean()
```

```python
avg_reading_score
```

```python
overall_avg_score = (avg_math_score + avg_reading_score) / 2
```

```python
overall_avg_score
```

```python
pass_math = school_data_complete['math_score'][school_data_complete['math_score']>= 70].count()
```

```python
pass_math
```

```python
pass_math_perc = pass_math / student_count * 100
```

```python
pass_math_perc
```

```python
pass_reading = school_data_complete['reading_score'][school_data_complete['reading_score']>= 70].count()
```

```python
pass_reading
```

```python
pass_reading_perc = pass_reading / student_count * 100
```

```python
pass_reading_perc
```

```python
district_summary = pd.DataFrame ({"Total Schools": [total_schools], "Total Students": [student_count], \
                                  "Total Budget": [total_budget], "Average Math Score": [avg_math_score], \
                                  "Average Reading Score": [avg_reading_score], "% Passing Math": [pass_math_perc], \
                                  " % Passing Reading": [pass_reading_perc], "% Overall Passing Rate": [overall_avg_score]})
```

```python
district_summary
```
